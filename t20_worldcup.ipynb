Import necessary libraries

```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
#len(data) = 1, so

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

We will use the scorecard as the unique ID

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv')
```

Create a match id dictionary

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
    
match_ids_dict
```

Batting Summary - One element in the json file, which contains all the match. Now each match has batting summary of each player for that match

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_records = []
    
    for rec in data:
        all_records.extend(rec['battingSummary']) # rec is recursive function

df_batting = pd.DataFrame(all_records)
df_batting.head()
```

Now clean the dismissal column, columns with details is OUT and null values are NOT OUT

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head()
```

Clean the bastmanName column

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(11)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
```

```python
df_batting.head()
```

```python
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv', index = False)
```

Bowling Summary

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv')
```
